```python
import datetime as dt

snapshot_date = df['ORDERDATE'].max()+dt.timedelta(days=1)

df_RFM = df.groupby(['CUSTOMERNAME']).agg({
    'ORDERDATE':lambda x:(snapshot_date-x.max()).days,
    'ORDERNUMBER':'count',
    'SALES':'sum'
})

df_RFM
```

```python
df_RFM.skew()
```

```python
df_RFM1 = np.log(df_RFM)
df_RFM1
```

```python
df_RFM1.skew()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
scaler.fit(df_RFM1)
```

```python
df_RFM1
```

```python
df_norm = scaler.transform(df_RFM1)
type(df_norm)
```

```python
df_norm = pd.DataFrame(df_norm,index=df_RFM1.index,columns=df_RFM1.columns)
df_norm
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

sse = {}
silhouette_avg = []

for k in range(2,21):
    kmeans = KMeans(n_clusters=k,random_state=1)
    kmeans.fit(df_norm)
    sse[k] = kmeans.inertia_
    cluster_labels = kmeans.labels_
    silhouette_avg.append(silhouette_score(df_norm,cluster_labels))

best_k = np.argmax(silhouette_avg)+2
best_k
```

```python
plt.plot(range(2,21),silhouette_avg)
```

```python
sns.lineplot(x=list(sse.keys()),y=list(sse.values()))
plt.xlabel('K')
plt.ylabel('SSE')
```

```python
kmeans_best = KMeans(n_clusters=3,random_state=1)
kmeans_best.fit(df_norm)

cluster_labels_best = kmeans_best.labels_

df_RFM['cluster'] = cluster_labels_best
df_RFM
```
